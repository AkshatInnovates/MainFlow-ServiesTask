```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Sales': [250, 270, 300, 320, 360, 400],
    'Expenses': [200, 180, 210, 240, 300, 330]
}

df = pd.DataFrame(data)
print(df)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Month'], df['Sales'], color='blue', label='Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['Sales'], marker='o', color='blue', label='Sales')
plt.plot(df['Month'], df['Expenses'], marker='o', color='red', label='Expenses')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Monthly Sales and Expenses')
plt.legend()
plt.show()
```

```python
# Bar Chart with Customizations
plt.figure(figsize=(10, 6))
plt.bar(df['Month'], df['Sales'], color='skyblue', label='Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.legend()
plt.grid(True)
plt.show()

# Line Chart with Customizations
plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['Sales'], marker='o', linestyle='-', color='blue', label='Sales')
plt.plot(df['Month'], df['Expenses'], marker='o', linestyle='--', color='red', label='Expenses')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Monthly Sales and Expenses')
plt.legend()
plt.grid(True)
plt.show()
```
